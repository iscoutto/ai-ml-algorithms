## k-Means utilizando Sklearn

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

## Base de dados :

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
#data
```

## Visualizando a base de dados :

```python
plt.scatter(data[0][:,0],data[0][:,1])#,c=data[1],cmap='rainbow')
```

## Algoritmo k-means com Sklearn :

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

## Centróides calculados :

```python
kmeans.cluster_centers_
```

## Clusters associados :

```python
kmeans.labels_
```

```python
kmeans.n_iter_
```

## Valor da função 'Elbow' para o valor de k escolhido:

```python
print('O valor da função Elbow para k =', kmeans.n_clusters, 'é', kmeans.inertia_)
```

## Visualização dos clusters associados :

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('k-Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
